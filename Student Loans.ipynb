```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.colors as pltcolors
import pandas as pd
import numpy as np
import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors as mcolors
from matplotlib import cm
import matplotlib.dates as mdates
from beautifultable import BeautifulTable
from datetime import datetime
from datetime import timedelta
import warnings
warnings.filterwarnings('ignore')

#######
#
# PLOT FUNCTIONS
#
#######

def plot_unemployment( unemployment ):

    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    #plot data
    x = unemployment[ 'Date' ]
    recent_graduates = unemployment[ 'Recent graduates' ]
    college_graduates = unemployment[ 'College graduates' ]
    all_workers = unemployment[ 'All workers' ]
    young_workers = unemployment[ 'Young workers' ]
    ###
    ax1.plot( x, recent_graduates, '-r', label='Recent Graduates' )
    ax1.plot( x, college_graduates, '-m', label='College Graduates' )
    #
    ax1.plot( x, all_workers, '-g', label='All Workers', linestyle='--' )
    ax1.plot( x, young_workers, '-b', label='Young Workers', linestyle='--' )

    ax1.legend( loc='center right' )
    fig.suptitle( 'Unemployment', fontsize=16 )
    vals = ax1.get_yticks()
    ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )
    
    years = mdates.YearLocator()   # every year
    months = mdates.MonthLocator()  # every month
    yearsFmt = mdates.DateFormatter( '%Y' )   

    #x-axis
    ax1.xaxis.set_major_locator( years )
    ax1.xaxis.set_major_formatter( yearsFmt )
    #ax1.xaxis.set_minor_locator( months )

    plt.subplots_adjust( top=.95 )
    plt.grid( linestyle='dotted' )
    
    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')
    
    plt.show()
    
def plot_underemployment( underemployment ):

    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    #plot data
    x = underemployment[ 'Date' ]
    recent_graduates = underemployment[ 'Recent graduates' ]
    college_graduates = underemployment[ 'College graduates' ]
    underemployment_all = underemployment[ 'underemployment_all' ]
    ###
    ax1.plot( x, recent_graduates, '-r', label='Underemployment, Recent Graduates' )
    ax1.plot( x, college_graduates, '-m', label='Underemployment, College Graduates' )
    ax1.plot( x, underemployment_all, '--c', label='Underemployment, All')
    
    plt.ylim([0, 50])

    ax1.legend( loc='center right' )
    fig.suptitle( 'Underemployment', fontsize=16 )
    vals = ax1.get_yticks()
    ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )
    
    years = mdates.YearLocator()   # every year
    months = mdates.MonthLocator()  # every month
    yearsFmt = mdates.DateFormatter( '%Y' )   

    #x-axis
    ax1.xaxis.set_major_locator(years)
    ax1.xaxis.set_major_formatter(yearsFmt)
    #ax1.xaxis.set_minor_locator(months)
    
    plt.subplots_adjust( top=.95 )
    plt.grid( linestyle='dotted' )
    
    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')
    plt.show()
    
def plot_enrollment( enrollment ):
    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    
    fig.suptitle( 'US Total Secondary Education Enrollment, 2005-2018', fontsize=16 )
    
    x = enrollment[ 'Year' ]
    
    ax1.plot( x, enrollment[ 'Total Enrollment' ] )
    ax1.set_ylabel('Total Enrollment' )
    plt.xticks( rotation=50 )
    
    ax2 = ax1.twinx()
    ax2.plot( x, enrollment[ 'Enrollment_pct_change' ], 'r--' )
    ax2.set_ylabel('Percent Change' )
    ax2.tick_params('y', colors='r')
    
    ax1.set_xticks( enrollment[ 'Year' ].tolist() )
    ax1.set_xticklabels( enrollment[ 'Year' ].tolist() )
    
    ax1.get_yaxis().set_major_formatter(
        ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    
    vals = ax2.get_yticks()
    ax2.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )
    
    cagr = ( (
        ( enrollment[ 'Total Enrollment' ].iloc[ -1 ] / enrollment[ 'Total Enrollment' ].iloc[ 0 ] ) ** 
        ( 1 / len( enrollment[ 'Total Enrollment' ].tolist() ) )
    ) - 1 ) * 100
    
    #annotate the CAGR
    ax1.annotate(
        '', 
        xy=( 
            enrollment[ 'Year' ].iloc[ -1 ], #x
            enrollment[ 'Total Enrollment' ].iloc[ -1 ] #y
        ),
        xytext=( 
            enrollment[ 'Year' ].iloc[ 0 ], #x
            enrollment[ 'Total Enrollment' ].iloc[ 0 ] #y
        ),
        arrowprops=dict( width=12, headwidth=18, facecolor='magenta' ) 
    )
    #annotate the CAGR
    ax1.text( 
        enrollment[ 'Year' ].iloc[ int( len( enrollment[ 'Year' ] ) / 2 ) ], 
        ( ( enrollment[ 'Total Enrollment' ].max() - enrollment[ 'Total Enrollment' ].min() ) / 2 + enrollment[ 'Total Enrollment' ].min() ),
        "CAGR: {:.2f}%".format( cagr ),
        fontsize=16,
        ha='center', 
        va='bottom'
    )

    plt.subplots_adjust( top=.95 )
    ax1.grid( linestyle='dotted' )
    plt.subplots_adjust( top=.95 )
    
    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')
    
    plt.show()
    
def plot_debt( debt, columns, callout=None, title=None, xaxis='Year', yaxistype='$' ):

    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    #plot data
    if xaxis == 'Year':
        x = pd.to_datetime( debt[ 'Year' ], format='%Y' )
    else:
        x = pd.to_datetime( debt[ xaxis ] )
    y = {}
    #get the Ys, which are the columns that were sent along
    for series in columns:
        y[ series ] = debt[ series]

    ###
    for key, value in y.items():
        if callout is None or callout not in key:
            ax1.plot( x, value, label='%s' % key )
        else:
            ax1.plot( x, value, linewidth=5, linestyle='--', label='%s' % ( key.split( "_" )[ 0 ] ) )

    plt.xticks(rotation=50)
    
    if title is not None:
        fig.suptitle( title, fontsize=16 )
    vals = ax1.get_yticks()
    if yaxistype == '$':
        ax1.set_yticklabels( [ '$%1.2fT' %i for i in vals ] )
        ax1.legend( loc='center right' )
    else:
        ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )
        plt.axhline(0, color='yellow', linewidth=4)
        ax1.legend( loc='top right' )
        
    years = mdates.YearLocator()   # every year
    months = mdates.MonthLocator()  # every month
    yearsFmt = mdates.DateFormatter( '%Y' )   

    #x-axis
    ax1.xaxis.set_major_locator( years )
    ax1.xaxis.set_major_formatter( yearsFmt )
    
    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')

    plt.subplots_adjust( top=.95 )
    plt.grid( linestyle='dotted' )
    plt.show()
    
def plot_cagr( data, columns, title ):
    cagr = {}

    for column in columns:
        cagr[ column ] = ( ( data[ column ].iloc[ -1 ] / data[ column ].iloc[ 0 ] )**( 1 / len( data[ column ].tolist() ) ) - 1 ) * 100
        
    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    fig.suptitle( title, fontsize=16 )
        
    plt.bar( cagr.keys(), cagr.values(), width=.5, color='cadetblue', edgecolor='black')
    
    vals = ax1.get_yticks()
    ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )
    
    plt.axhline(0, color='black', linewidth=2)
    
    #create labels
    rects = ax1.patches

    # Make some labels.
    for rect, label in zip( rects, cagr.values() ):
        height = rect.get_height()
        thisheight = ( height - 0.4 ) if label < 0 else ( height + 0.2 )
        ax1.text( rect.get_x() + rect.get_width() / 2, thisheight, 
            "{0:.2f}".format( label ) + "%", ha='center', va='bottom')
        
    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')
    
    plt.subplots_adjust( top=.95 )
    plt.grid( linestyle='dotted' )
    plt.show()
    
def plot_enrollment_vs_student_debt( enrollment, debt_yearly ):
    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    
    fig.suptitle( 'US Total Secondary Education Enrollment vs. Student Loan Debt, 2005-2018', fontsize=16 )
    
    x = enrollment[ 'Year' ]
    
    ax1.plot( x, enrollment[ 'Total Enrollment' ], linewidth=4, linestyle='--', label='Total Enrollment')
    ax1.set_ylabel('Total Enrollment' )
    plt.xticks( rotation=50 )
    
    ax2 = ax1.twinx()
    ax2.plot( x, debt_yearly[ 'Student Loan' ], 'm', linewidth=4, label='Student Loan Debt' )
    ax2.set_ylabel('Student Loan Debt' )
    ax2.tick_params('y', colors='r')
    
    ax1.set_xticks( enrollment[ 'Year' ].tolist() )
    ax1.set_xticklabels( enrollment[ 'Year' ].tolist() )
    
    ax1.get_yaxis().set_major_formatter(
        ticker.FuncFormatter( lambda x, p: format( int( x ), ',' ) ) )

    vals = ax2.get_yticks()
    ax2.set_yticklabels( [ '$%1.2fT' %i for i in vals ] )
    
    fig.legend(loc=1, bbox_to_anchor=(1,0.5), bbox_transform=ax1.transAxes)

    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')

    plt.subplots_adjust( top=.95 )
    ax1.grid( linestyle='dotted' )
    plt.show()
    
def plot_debt_delinquency_vs_underemployment( delinquent_debt, uq, unq ):
    
    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    
    fig.suptitle( 'US Student Loan Delinquency Rates vs. Unemployment & Underemployment Rates, 2005-2018', fontsize=16 )
    
    x = delinquent_debt[ 'Date' ]
    
    plt.xticks( rotation=50 )
    
    ax1.plot( x, delinquent_debt[ 'Student Loan Delinquent Percent' ], color='magenta',
        linewidth=4, linestyle='-', label='Delinquent Student Loan Debt (%)')
    ax1.set_ylabel('Percentage of Student Loan Debt in Delinquency' )
    
    #underemployment
    ax2 = ax1.twinx()
    ax2.plot( x, unq[ 'Recent graduates' ], color='teal', linewidth=3, linestyle='-.', label='Underemployment of Recent Graduates' )
    ax2.plot( x, unq[ 'College graduates' ], color='darkolivegreen', linewidth=3, linestyle='-.', label='Underemployment of All College Graduates' )
    ax2.set_ylabel('Unemployment & Underemployment Rate' )
    ax2.tick_params('y', colors='r')
    
    #unemployment
    ax2.plot( x, uq[ 'Recent graduates' ], color='darkslategray', linewidth=3, linestyle='-.', label='Unemployment of Recent Graduates' )
    
    ax1.set_xticks( delinquent_debt[ 'Date' ].tolist() )
    ax1.set_xticklabels( delinquent_debt[ 'Date' ].tolist() )
    for label in ax1.xaxis.get_ticklabels()[1::2]:
        label.set_visible( False )
    
    vals = ax1.get_yticks()
    ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )

    vals = ax2.get_yticks()
    ax2.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )

    fig.legend(loc=1, bbox_to_anchor=(1,0.5), bbox_transform=ax1.transAxes)
    
    fig.text(.5, .025, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')

    plt.subplots_adjust( top=.95 )
    ax1.grid( linestyle='dotted' )
    plt.show()
    
def plot_debt_delinquency_vs_enrollment( dy, enrollment ):
    
    fig, ax1 = plt.subplots()
    plt.rcParams[ 'figure.figsize' ] = [ 12, 8 ]
    
    fig.suptitle( 'US Student Loan Delinquency Amounts vs. Secondary Education Enrollment, 2005-2018', fontsize=16 )
    
    x = dy[ 'Year' ]
    
    plt.xticks( rotation=50 )
    
    ax1.plot( x, dy[ 'Student Loan Delinquent Percent' ], color='magenta',
        linewidth=4, linestyle='-', label='Delinquent Student Loan Debt (%)')
    ax1.set_ylabel('Percentage of Student Loan Debt in Delinquency' )
    
    #enrollment
    ax2 = ax1.twinx()
    ax2.plot( x, enrollment[ 'Total Enrollment' ], color='teal', linewidth=3, linestyle='-.', label='Total Enrollment' )
    ax2.set_ylabel('Total Secondary Education Enrollment' )
    ax2.tick_params('y', colors='r')
    
    ax1.set_xticks( dy[ 'Year' ].tolist() )
    ax1.set_xticklabels( dy[ 'Year' ].tolist() )
    
    vals = ax1.get_yticks()
    ax1.set_yticklabels( [ '%1.2f%%' %i for i in vals ] )

    ax2.get_yaxis().set_major_formatter(
        ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

    fig.legend(loc=1, bbox_to_anchor=(1,0.5), bbox_transform=ax1.transAxes)

    fig.text(.5, .05, "Data: NY and St. Louis Fed, National Center for Education Statistics. Created by Eric Benjamin Seufert (@eric_seufert)", ha='center')
    
    plt.subplots_adjust( top=.95 )
    ax1.grid( linestyle='dotted' )
    plt.show()
    
#####
#
# END PLOT FUNCTIONS
#
#####

#######
#
# DATA PREP FUNCTIONS
#
#######
    
def prep_unemployment_data():
    #unemployment
    unemployment = pd.read_excel( 
        open( 'labormarket_grads.xlsx', 'rb' ), sheet_name='ch1_unemployment', 
        header=13, skipfooter=2, parsedates=[ 1 ] )
    #filter it from 2000 on
    unemployment = unemployment[ ( unemployment[ 'Date' ] >= '2005-01-01' ) ]

    #underemployment
    underemployment = pd.read_excel( 
        open( 'labormarket_grads.xlsx', 'rb' ), sheet_name='ch2_underemployment', 
        header=14, skipfooter=2, parsedates=[ 1 ] )
    underemployment = underemployment[ ( underemployment[ 'Date' ] >= '2005-01-01' ) ]

    #unemployment including all underemployment
    underemployment_all = pd.read_excel( 
        open( 'civ_unemployment_plus_discouraged.xls', 'rb' ), header=None, 
        skiprows=11, parsedates=[ 1 ] )
    underemployment_all.columns = [ 'Date', 'Unemployment Rate with Underemployment' ]
    underemployment_all = underemployment_all[ ( underemployment_all[ 'Date' ] >= '2005-01-01' ) ]

    #civilian unemployment baseline, to be subtracted from above
    civ_unemployment = pd.read_excel( 
        open( 'civilian_unemployment.xls', 'rb' ), header=None, 
        skiprows=11, parsedates=[ 1 ] )
    civ_unemployment.columns = [ 'Date', 'Unemployment Rate' ]
    civ_unemployment = civ_unemployment[ ( civ_unemployment[ 'Date' ] >= '2005-01-01' ) ]

    #combined dataframes from above
    civ_un_underemployment = civ_unemployment.set_index( 'Date' ).join( underemployment_all.set_index( 'Date' ), on='Date', how='left', lsuffix="_review" )
    civ_un_underemployment[ 'underemployment_all' ] = ( civ_un_underemployment[ 'Unemployment Rate with Underemployment' ]
        - civ_un_underemployment[ 'Unemployment Rate' ] )

    #join the underemployment_all data to the underemployment for young people data
    underemployment = underemployment.join( civ_un_underemployment,
        on='Date', how='left', lsuffix="_review" )    
    
    return unemployment, underemployment

"""
def prep_quarterly_debt_data():
    #debt load
    debt = pd.read_excel( 
        open( 'HHD_C_Report_2018Q4.xlsx', 'rb' ), header=2, skiprows=1, parsedates=[ 1 ], 
        sheet_name='Page 3 Data'
    )
    #get rid of the extraneous columns
    debt = debt.iloc[ :,0:8 ]
    #change the first column to be named Date
    debt = debt.rename( columns = { 'Unnamed: 0': 'Date_Quarters' } )
    
    #parse the Date strings
    quarter_dates = [ "20" + x.split( ":" )[ 0 ] + "-" + x.split( ":" )[ 1 ] for x in debt[ 'Date_Quarters' ].tolist() ]
    years = [ int( "20" + x.split( ":" )[ 0 ] ) for x in debt[ 'Date_Quarters' ].tolist() ]
    debt[ 'Date' ] = quarter_dates
    debt[ 'Year' ] = years
    debt.set_index( 'Date' )
    debt = debt[ debt[ 'Year' ] >= 2005 ]
    
    return debt
"""

def prep_yearly_debt_data( debt_quarterly ):
    #debt_columns = ['Year', 'Mortgage', 'HE Revolving', 'Auto Loan', 'Credit Card', 'Student Loan', 'Other', 'Total' ]
    debt_yearly = debt_quarterly[ debt_quarterly.columns.tolist() ].groupby( [ 'Year' ] ).sum()
    debt_yearly.reset_index( level=0, inplace=True )

    for column in [ c for c in debt_quarterly.columns.tolist() if c not in [ "Year", "Total", "Date_Quarters", "Quarter", "Date" ] ]:
        debt_yearly[ column + '_pct_change' ] = debt_yearly[ column ].pct_change() * 100
        
    return debt_yearly

def prep_enrollment_data():
    enrollment = pd.read_excel( 
        open( 'postsecondary_enrollment.xls', 'rb' ), header=2, skiprows=1, parsedates=[ 1 ]
    )
    #get rid of the extraneous columns
    enrollment = enrollment[ 1:]
    #only care about the first two columns, year and total enrollment
    enrollment = enrollment.iloc[ :,0:2 ]
    #name the columns
    enrollment = enrollment.rename( columns = { 'Unnamed: 0': 'Year', 'Unnamed: 1': 'Total Enrollment' } )
    #remove the NaN
    enrollment = enrollment.dropna()
    #fix the year formatting
    enrollment[ 'Year' ] = enrollment[ 'Year' ].apply( lambda x: int( str( x ).split('\\')[ 0 ][ :4 ] ) )
    #only care about 2005-2018
    enrollment = enrollment[ ( enrollment[ 'Year' ] >= 2005 ) & ( enrollment[ 'Year' ] <= 2018 ) ]
    #add a percent change column
    enrollment[ 'Enrollment_pct_change' ] = enrollment[ 'Total Enrollment' ].pct_change() * 100
    
    #cagr = ( ( enrollment[ 'Total Enrollment' ].iloc[ -1 ] / enrollment[ 'Total Enrollment' ].iloc[ 0 ] ) ** ( 1 / len( enrollment[ 'Total Enrollment' ].tolist() ) ) - 1 ) * 100
    
    return enrollment

def prep_quarterly_debt_data():
    #debt load
    debt = pd.read_excel( 
        open( 'HHD_C_Report_2018Q4.xlsx', 'rb' ), header=2, skiprows=1, parsedates=[ 1 ], 
        sheet_name='Page 3 Data'
    )
    #get rid of the extraneous columns
    debt = debt.iloc[ :,0:8 ]
    #change the first column to be named Date
    debt = debt.rename( columns = { 'Unnamed: 0': 'Date_Quarters' } )
    
    #parse the Date strings
    #debt[ 'Date' ] = debt[ 'Date_Quarters' ].str.split( pat = ":" )[ 1 ] + " " + "20" + debt[ 'Date_Quarters' ].str.split( pat = ":" )[ 0 ]
    quarter_dates = [ "20" + x.split( ":" )[ 0 ] + "-" + x.split( ":" )[ 1 ] for x in debt[ 'Date_Quarters' ].tolist() ]
    years = [ int( "20" + x.split( ":" )[ 0 ] ) for x in debt[ 'Date_Quarters' ].tolist() ]
    debt[ 'Quarter' ] = debt[ 'Date_Quarters' ].apply( lambda x: str( x.split( ":" )[ 1 ] ) )
    debt[ 'Date' ] = quarter_dates
    debt[ 'Date' ] = debt[ 'Date' ].astype( str )
    debt[ 'Year' ] = years
    debt = debt[ debt[ 'Year' ] >= 2005 ]
    
    debt.reset_index( level=0, inplace=True )
    debt.set_index( 'Date' )

    return debt

def prep_delinquent_debt_data( debt_quarterly ):
    #delinquent debt
    delinquent_debt = pd.read_excel( 
        open( 'HHD_C_Report_2018Q4.xlsx', 'rb' ), header=2, skiprows=1, 
        sheet_name='Page 12 Data'
    )
    delinquent_debt = delinquent_debt.iloc[ :,0:8 ]
    #change the first column to be named Date
    delinquent_debt = delinquent_debt.rename( 
        columns = { 'Unnamed: 0': 'Date', 'MORTGAGE': 'Mortgage Delinquent Percent',
        'HELOC': 'HELOC Delinquent Percent', 'AUTO': 'Auto Loan Delinquent Percent',
        'CC': 'Credit Card Delinquent Percent', 'STUDENT LOAN': 'Student Loan Delinquent Percent',
        'OTHER': 'Other Delinquent Percent', 'ALL': 'All Delinquent Percent' } )

    #parse the Date strings
    #for some reason about 1/2 of the first column in Page 12 Data gets interpreted as a datetime object
    #vs a string for Page 3 Data. Only need to extract Year from the datetime object
    delinquent_debt[ 'Year' ] = delinquent_debt[ 'Date' ].apply( lambda x: x.year if isinstance( x, datetime ) else int( "20" + x.split( ":" )[ 0 ] ) )
    delinquent_debt[ 'Quarter' ] = delinquent_debt[ 'Date' ].apply( lambda x: "Q" + str( pd.Timestamp( x ).quarter ) if isinstance( x, datetime ) else str( x.split( ":" )[ 1 ] ) )
    
    #now reconstruct the date
    delinquent_debt = delinquent_debt.drop( columns = [ 'Date' ] )
    delinquent_debt[ 'Date' ] = delinquent_debt.apply( lambda row: str( str( row.Year ) + "-" + str( row.Quarter ) ), axis=1 )
    delinquent_debt[ 'Date' ] = delinquent_debt[ 'Date' ].astype( str )
    
    delinquent_debt = delinquent_debt[ delinquent_debt[ 'Year' ] >= 2005 ]
    
    delinquent_debt.reset_index( level=0, inplace=True )
    delinquent_debt.set_index( 'Date' )
    
    #now join the delinquent debt amounts to the quarterly debt %s to get absolute numbers
    delinquent_debt = delinquent_debt.join( debt_quarterly, how = 'left', lsuffix = '_l' )
    #remove all the overlapping columns
    delinquent_debt = delinquent_debt[ [ c for c in delinquent_debt.columns.tolist() if '_l' not in c ] ]
    
    return delinquent_debt

def transform_underemployment_data_quarterly( underemployment ):
    #transform underemployment into a list of the underemployment rates by year for December
    underemployment[ 'Year' ] = underemployment[ 'Date' ].apply( lambda x: int( x.date().year ) )
    underemployment[ 'Month' ] = underemployment[ 'Date' ].apply( lambda x: int( x.date().month ) ) 
    uq = underemployment[ underemployment[ 'Month' ].isin( [ 3, 6, 9, 12 ] ) ]
    uq[ 'Quarter' ] =  ( underemployment[ 'Month' ] / 3 ).astype( int )
    uq[ 'Date' ] = uq.apply( lambda row: str( str( row.Year ) + "-" + "Q" + str( row.Quarter ) ), axis=1 )
    
    return uq

def transform_unemployment_data_quarterly( unemployment ):
    #transform unemployment
    unemployment[ 'Year' ] = unemployment[ 'Date' ].apply( lambda x: int( x.date().year ) )
    unemployment[ 'Month' ] = unemployment[ 'Date' ].apply( lambda x: int( x.date().month ) ) 
    unq = unemployment[ unemployment[ 'Month' ].isin( [ 3, 6, 9, 12 ] ) ]
    unq[ 'Quarter' ] =  ( unemployment[ 'Month' ] / 3 ).astype( int )
    unq[ 'Date' ] = unq.apply( lambda row: str( str( row.Year ) + "-" + "Q" + str( row.Quarter ) ), axis=1 )
    
    return unq

def transform_delinquent_debt_data_yearly( delinquent_debt ):
    #transform delinquent debt to be yearly
    dy = dy[ [ 'Mortgage', 'HE Revolving', 'Auto Loan', 'Credit Card', 'Student Loan', 
        'Other', 'Total', 'Year' ] ]
    
    dy = delinquent_debt.groupby( 'Year' ).sum()
    print( dy )

def transform_delinquent_debt_data_yearly( delinquent_debt ):
    #get the Q4 value for each year
    delinquent_debt = delinquent_debt[ delinquent_debt[ 'Quarter' ].isin( [ 'Q4' ] ) ]
    delinquent_debt[ 'Student Loan Delinquent Absolute' ] = delinquent_debt.apply( lambda row: float( row[ 'Student Loan Delinquent Percent' ] ) / 100 * ( float( row[ 'Student Loan' ] ) ), axis = 1 ) 
    #transform delinquent debt to be yearly
    
    dy = delinquent_debt[ [ 'Student Loan Delinquent Absolute', 'Student Loan Delinquent Percent', 'Year' ] ]
    
    return dy
    
unemployment, underemployment = prep_unemployment_data()
debt_quarterly = prep_quarterly_debt_data()
debt_yearly = prep_yearly_debt_data( debt_quarterly )
delinquent_debt = prep_delinquent_debt_data( debt_quarterly )
enrollment = prep_enrollment_data()
quarterly_unemployment = transform_unemployment_data_quarterly( unemployment )
quarterly_underemployment = transform_underemployment_data_quarterly( underemployment )
yearly_delinquent_debt = transform_delinquent_debt_data_yearly( delinquent_debt )
    
#plot_unemployment( unemployment )
#plot_underemployment( underemployment )
#plot_debt( debt_yearly, [ c for c in debt_yearly.columns.tolist() if c not in [ "Year", "Total" ] ], callout="Student Loan", title="US Household Debt" )
#plot_debt( debt_yearly, [ c for c in debt_yearly.columns.tolist() if c not in [ "Year", "Total", "Mortgage" ] ], callout="Student Loan", title="US Household Debt (excluding Mortgage Debt) " )
#plot_debt( debt_yearly, [ c for c in debt_yearly.columns.tolist() if c not in [ "Year" ] and "_pct_change" in c ], callout="Student Loan", title="US Household Debt Growth Rates", yaxistype='%' )
#plot_cagr( debt_yearly, [ c for c in debt_yearly.columns.tolist() if c not in [ "Year", "Total" ] ], "CAGR of US Household Debt Types from %i to %i" %( min( debt_yearly[ 'Year'] ), max( debt_yearly[ 'Year' ] ) ) )

#plot_debt( delinquent_debt, [ c for c in delinquent_debt.columns.tolist() if c not in [ "YEAR", "QUARTER", "ALL" ] and "Delinquent" in c ], callout="Student Loan Delinquent Percent", title="FDSFDSFS", xaxis='Date', yaxistype='%' )
#plot_enrollment_vs_student_debt( enrollment, debt_yearly )
#plot_debt_delinquency_vs_underemployment( delinquent_debt, quarterly_unemployment, quarterly_underemployment )
#plot_debt_delinquency_vs_enrollment( yearly_delinquent_debt, enrollment )

"""
"""
```
